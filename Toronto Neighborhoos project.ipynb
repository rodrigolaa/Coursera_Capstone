1) read the tables in the Wikipedia page with the read_html command. 

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

2) Analyze the table dimensions 

```python
df_nbh = df[0]
df_nbh.shape
```

3) Select all the rows that doesn't have 'Not assigned' in the column Borough

```python
df_toronto = df_nbh[df_nbh.Borough != 'Not assigned']
df_toronto.shape
df_toronto
    
```

4) Show what rows have 'Not assigned' in the column Neighbourhood

```python
df_toronto[df_toronto.Neighbourhood == 'Not assigned']
```

5) Substitute the cell with the name of the Borough

```python
df_toronto.at[8,'Neighbourhood'] = df_toronto['Borough'][8]
df_toronto.loc[df_toronto.index == 8]
```

6) Group all the Neighbouhoods with the same Postcode in the same row

```python
df_toronto['Postcode'].unique().shape
```

```python
df_toronto = df_toronto.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda x: ','.join(x.astype(str))).reset_index()
```

7) Now import the csv file with the coordinates and merge it with the df_toronto

```python
df_coor = pd.read_csv('http://cocl.us/Geospatial_data', index_col = False, skiprows = 0)
df_coor
```

```python
#It is necessary to change the column name of the Postal codes to be exactly as the df_toronto because it is the column id to merge both dataframes
df_coor = df_coor.rename(columns = {'Postal Code': 'Postcode'})
```

```python
df_toronto_coor = pd.merge(df_toronto, df_coor, on = 'Postcode')
df_toronto_coor
```

8) Create a Folium map to show all the boroughs in toronto

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto_coor['Latitude'], df_toronto_coor['Longitude'], df_toronto_coor['Borough'], df_toronto_coor['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
